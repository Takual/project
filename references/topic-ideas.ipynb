## Analyse von Strom- und Gastarifen in Deutschland

### Data source

Der Datensatz umfasst sämtliche Strom- und Gastarife aller Versorger in Deutschland. Die Daten stammen von einem Anbieter, der umfassende Tarifinformationen für den Stichtag 1. Januar 2025 bereitstellt. Zusätzlich werden historische Daten zum Vergleich mit Anfang 2024 integriert. Mögliche Ergänzungen stammen aus öffentlichen Quellen, etwa Verbrauchsdaten, soziodemografischen Statistiken oder Angaben zur Energieerzeugung.

### Data characteristics

 Der Datensatz enthält folgende Hauptmerkmale:

Tarifarten: Grundversorgung und Sondertarife, differenziert nach Haushaltsgrößen (Singlehaushalt/4-Personenhaushalt) und Verbrauchsmengen.
Regionaldaten: Postleitzahlgebiete und die zugehörigen Anbieter, wobei mehrere Anbieter in einem Postleitzahlgebiet aktiv sein können.
Zeitbezug: Daten zum Stichtag 1. Januar 2025 sowie Vergleichsdaten vom Anfang des Jahres 2024.
Ergänzende Daten: Verbrauchsstatistiken, Soziodemografische Daten (z. B. Altersstruktur, Einkommen), Wettbewerbsdaten (Anzahl Anbieter pro Region), und Informationen zur Energieerzeugung (z. B. Anteil erneuerbarer Energien).

### Research question
Die Analyse zielt darauf ab, verschiedene Forschungsfragen zu beantworten:

Vergleich von Grundtarifen und Sonderverträgen:

Ziel: Untersuchung der Preisunterschiede zwischen Grundtarifen und Sonderverträgen.
Ansatz: Analyse durchschnittlicher Preise, Preisspannen und Verteilungen für beide Tarifarten.
Einflussfaktoren auf die Preise:

Ziel: Identifikation der Faktoren, die Strom- und Gaspreise beeinflussen.
Prädiktoren: Tarifart (Grundtarif vs. Sondervertrag), Verbrauch, Region, Art der Energieerzeugung (z. B. Ökostrom), Anbietergröße.
Methodik: Regressionsanalyse zur Quantifizierung des Einflusses einzelner Faktoren.
Kundensegmentierung:

Ziel: Identifikation von Kundengruppen mit ähnlichen Merkmalen und Präferenzen.
Methodik: Clusteranalyse zur Segmentierung der Kunden für optimierte Angebotserstellung.
Preisoptimierung:

Ziel: Entwicklung eines Modells zur Empfehlung des optimalen Tarifs basierend auf individuellem Verbrauch und Bedürfnissen.
Ansatz: Aufbau eines Empfehlungssystems mit den bereitgestellten Daten.
Zusätzliche Analysen:

Bereitstellung von Daten und Insights für journalistische Berichterstattung.
Erstellung einfacher grafischer Elemente zur Verbraucherinformation.
Entwicklung eines Dashboards für interaktive Datenanalyse.

## Analyse der Kraftstoffpreise an deutschen Tankstellen: Einfluss der Raffinerieentfernung und anderer Faktoren

### Data source

Die primäre Datenquelle ist ein umfassender Datensatz, der die täglichen Kraftstoffpreisänderungen aller Tankstellen in Deutschland über die letzten 10 Jahre dokumentiert. Dieser Datensatz enthält Informationen wie Name, Adresse, geografische Koordinaten und Kettenzugehörigkeit der Tankstellen.

Zusätzlich werden folgende Datenquellen genutzt:

Geodaten von Raffinerien: Zur Berechnung der Entfernung zwischen Tankstellen und der nächstgelegenen Raffinerie.
Sektoranalyse des Bundeskartellamts: Informationen über die regionale Marktdominanz und die Versorgungsketten der Raffinerien.
Sekundärdaten wie Rohölpreise (z. B. von der EIA), Wetterbedingungen (über historische Wetter-APIs) sowie Verkehrs- und Feiertagsdaten.


### Data characterisitcs

Beobachtungen: Tägliche Preisaufzeichnungen von mehreren tausend Tankstellen über 10 Jahre, was zu deutlich mehr als 100 Beobachtungen führt.
Variablen: Der Datensatz enthält über 10 Spalten, darunter:
Tankstellendetails: Name, Adresse, Geokoordinaten, Kettenzugehörigkeit.
Tägliche Preisänderungen: Preise für Benzin und Diesel.
Geografische Merkmale: Entfernung zur nächstgelegenen Raffinerie (berechnet aus den Koordinaten).
Marktbedingungen: Wettbewerb in der Umgebung (z. B. Anzahl nahegelegener Tankstellen).
Externe Faktoren: Rohölpreis, Wetterdaten, Verkehrsdichte und Feiertage.
Mindestens sechs unabhängige Prädiktorvariablen (z. B. Raffinerieentfernung, Wettbewerbsdichte, Kettenzugehörigkeit) sind vorhanden. Die abhängigen Variablen umfassen Kraftstoffpreise für Regression und Preiskategorien für Klassifikation. Die Daten sind tabellarisch strukturiert und erfüllen die Unabhängigkeitsannahme, da sie tägliche Preisänderungen und keine wiederholten Messungen darstellen.

### Research question

Zentrale Forschungsfragen:

Regressionsanalyse:

Welchen Einfluss hat die Entfernung zur nächstgelegenen Raffinerie auf die Kraftstoffpreise an Tankstellen?
Welche externen Faktoren wie Verkehrsaufkommen, Rohölpreise oder Wettbewerbsdichte beeinflussen die Preisgestaltung?
Klassifikationsanalyse:

Können Tankstellen basierend auf ihrer Nähe zu Raffinerien, ihrer Preisentwicklung und anderen Merkmalen in "günstig", "durchschnittlich" und "teuer" klassifiziert werden?
Gibt es regionale Muster, die auf Unterschiede in den Preisstrategien hinweisen?
Abhängige Variablen:

Für Regression: Kraftstoffpreise (kontinuierlich).
Für Klassifikation: Preiskategorien (kategorial: günstig, durchschnittlich, teuer).
Unabhängige Variablen:

Entfernung zur Raffinerie.
Kettenzugehörigkeit (z. B. Marke oder unabhängige Tankstelle).
Anzahl der Wettbewerber in der Umgebung.
Rohölpreise.
Verkehrsaufkommen in der Region.
Wetterbedingungen (z. B. Regen, Temperatur).
Wochentag und Feiertage.
